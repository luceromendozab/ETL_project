```python
import pandas as pd
from IPython.display import display
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

## Around the world

```python
df= pd.read_csv("../clean_data/df_languages.csv", index_col= 0)
df.head(4)
```

### Endargement 

```python
# to see  how many languages are endangered arpund the world 

plt.figure(figsize = (8,4)) 

plt.hist(df["endangerment"], edgecolor = "black", 
                        facecolor = "#0072BD", 
                        bins = 10)
plt.xticks(rotation=45);
```

## How many languages by country 

```python
counts= df["countries"].value_counts()
```

```python
#top 10 countries with endangered languages 
counts_sort = counts.sort_values(ascending=False)
counts_sort.head(10)
```

## US

```python
df_2= pd.read_csv("../clean_data/df_usa_languages.csv", index_col= 0)
df_2.head(4)
```

```python
# to see  how many languages are in US

plt.figure(figsize = (8,4)) 

plt.hist(df_2["endangerment"], edgecolor = "black", 
                        facecolor = "#0072BD", 
                        bins = 10)
plt.xticks(rotation=45);
```

```python
df_more_speakers = df_2[df_2["speakers"]> 10000]
df_more_speakers.head()
```

```python
# languages with more speakers
ax = sns.barplot(x = 'language', y = 'speakers',data = df_more_speakers)
ax.set_xticklabels(ax.get_xticklabels(),rotation = 30)
```

## Us population

```python
df_3= pd.read_csv("../clean_data/df_usa_pop.csv", index_col= 0)
df_3.head(4)
```

```python
#Population since 1970 until 1990 by state 
fig1 = df_3.plot(x='state', y=['1970', '1980','1990'], 
              figsize=(20,7),  linewidth=0.7, marker= ".")
```

```python
#Population since 2000 until 2020 by state
fig2 = df_3.plot(x='state', y=['2000', '2010','2020'], 
              figsize=(20,7),  linewidth=0.7, marker= ".")
```
