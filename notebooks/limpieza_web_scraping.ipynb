```python
import pandas as pd
import numpy as np
```

```python
df_pop= pd.read_csv("df_population_usa.csv", index_col=0)
df_pop.head(5)
```

```python
df_pop.columns
```

```python
df_pop=df_pop.replace('\n','', regex=True)
df_pop.head()
```

```python
df_1=df_pop.copy()
df_1.head()
```

```python
df_1.columns
```

```python
df_1.columns[1:9]
```

```python
columns_to_convert= df_1.columns[1:9]
```

```python
df_1=df_1.drop(df_1.index[-1])
```

```python
for col in columns_to_convert:
    df_1[col] = df_1[col].replace( '[^0-9]', '', regex=True )        
```

```python
df_1
```

```python
# Iterate through the list of column names
for col in columns_to_convert:
    # Check if the column name exists in the DataFrame
    df_1[col] = df_1[col].apply(lambda x: int(x))
```

```python
df_1.info()
```
