```python
import pandas as pd
import requests
import json
```

```python
#Llamada a la API para obtener el numero de población 

response = requests.get('http://api.worldbank.org/v2/country/US/indicator/SP.POP.TOTL?format=json')
data = json.loads(response.content)
```

```python
response.status_code
```

```python
data_d= data.pop(1)
```

```python
#data_d
```

```python
df_pop=pd.DataFrame(data_d)
df_pop.head(2)
```

```python
df_pop.info()
```

```python
df_pop.columns
```

```python
# datos del pib de USA de los ultimos 5 años 
df_new=df_pop.drop(labels=['indicator','country','unit','obs_status','decimal'], axis=1)
df_new.head(10)
```

```python
df_new = df_new.rename(columns={'countryiso3code':'country',
                                   'value':'population'})
df_new.head(1)
```
