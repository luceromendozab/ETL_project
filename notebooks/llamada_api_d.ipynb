## Es el cociente entre el crecimiento natural anual (nacimientos menos defunciones) de un determinado perodo y la poblacin media del mismo perodo. 



Por ejemplo, para obtener los 
datos de densidad de poblacin de un estado especfico en el ao 2019, la URL sera algo as:
    
    
    https://api.census.gov/data/2019/pep/population?get=NAME,DENSITY&for=STATE:XX
    
Tambin puedes acceder a los datos histricos de poblacin y densidad de poblacin para todos los estados de Estados Unidos desde los aos 1910 hasta 2019.

https://api.census.gov/data/timeseries/pep/population
El ao de los datos que deseas obtener, utilizando el parmetro "time". Por ejemplo, "time=2010" para obtener los datos de 2010.

El nivel geogrfico para los datos que deseas obtener, utilizando el parmetro "for". Por ejemplo, "for=STATE:*" para obtener los datos para todos los estados.

Los campos que deseas obtener, utilizando el parmetro "get". Por ejemplo, "get=NAME,POP,DENSITY" para obtener el nombre del estado, la poblacin y la densidad de poblacin

-------------

https://api.census.gov/data/timeseries/pep/population?get=NAME,POP,DENSITY&time=from+1910&for=STATE:*&key=YOUR_API_KEY

https://api.census.gov/data/timeseries/pep/population es la URL base para acceder a los datos histricos de poblacin y densidad de poblacin.
get=NAME,POP,DENSITY especifica los campos que deseas obtener, en este caso el nombre del estado, la poblacin y la densidad de poblacin.
time=from+1910 especifica el rango de tiempo para los datos, en este caso desde el ao 1910 hasta la fecha actual.
for=STATE:* especifica el nivel geogrfico para los datos, en este caso todos los estados.
key=YOUR_API_KEY especifica tu clave de acceso a la API.


Composicin tnica de la poblacin
natalidad, mortalidad migracion 

Para obtener estos datos, debes especificar los campos que deseas obtener en el parmetro "get" en la URL de la API. Por ejemplo, si quieres obtener datos sobre la natalidad, debes incluir el campo "BIRTHS" en el parmetro "get".

https://api.census.gov/data/timeseries/pep/population?get=NAME,BIRTHS&time=from+1910&for=STATE:*&key=YOUR_API_KEY

"NAME" (nombre del estado)
"NAME_WHITE" (poblacin blanca)
"NAME_BLACK" (poblacin negra)
"NAME_AMIN" (poblacin indgena americana)
"NAME_ASIAN" (poblacin asitica)
"NAME_NHOPI" (poblacin nativa hawaiana y de otras islas del Pacfico)
"NAME_OTHER" (poblacin de otra raza)
"NAME_2MORE" (poblacin de dos o ms razas)

https://api.census.gov/data/timeseries/pep/population?get=NAME,NAME_WHITE,NAME_BLACK&time=from+1910&for=STATE:*&key=YOUR_API_KEY
